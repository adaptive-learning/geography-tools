## Response Time

```python
def plot_time_hist():
    figure(figsize=(15,5))
    xmax = max(answers.wrong_answers()["response.time"].max(), answers.correct_answers()["response.time"].max())
    plt.subplot(1, 2, 1)
    plt.hist(answers.correct_answers()["response.time"],
             bins = 20,
             histtype = "step",
             color = "green",
             label = "Correct Answers")
    plt.legend()
    plt.xlabel("Response Time (log)")
    plt.xlim(0, xmax)
    plt.subplot(1, 2, 2)
    plt.hist(answers.wrong_answers()["response.time"],
             bins = 20,
             histtype = "step",
             color = "red",
             label = "Wrong Answers")
    plt.legend()
    plt.xlabel("Response Time (log)")
    plt.xlim(0, xmax)
    plt.show()

plot_time_hist()
```

## Probability of Correct Answer Based on Previous Answers

```python
def plot_probs(x, y, lbl, xlbl = "Number of Answers Before"):
    plt.ylim(0, 1.1)
    plt.plot(x, y, label = lbl)
    plt.ylabel("Probability Next Answer Is Correct")
    plt.xlabel(xlbl)
    plt.legend(loc = 4)

def plot_prob_correct_answer_by_column(column, label):
    to_plot = pd.DataFrame()
    max_num_of_answers = max(answers.data[column])
    for n in range(1, max_num_of_answers + 1):
        filtered = answers.data[answers.data[column] == n]
        if len(filtered) == 0:
            continue
        correct = filtered[filtered["place.asked"] == filtered["place.answered"]]
        to_plot = to_plot.append(pd.DataFrame({"number" : [n],
                                               "prob"   : [len(correct)/float(len(filtered))]}))
    plot_probs(to_plot["number"], to_plot["prob"], label)
    to_plot = pd.DataFrame()
    
def plot_prob_correct_answer_by_success():
    ratios = pd.DataFrame()
    for i, row in answers.data.iterrows():
        if not row["question.number.all"]:
            continue
        before = row["question.number.correct"] / float(row["question.number.all"])
        ratios = ratios.append(pd.DataFrame({"ratio"   : [before],
                                    "correct" : [row["place.asked"] == row["place.answered"]]}))
    to_plot = pd.DataFrame()
    for r in ratios["ratio"].unique():
        filtered = ratios[ratios["ratio"] == r]
        to_plot = to_plot.append(pd.DataFrame({"ratio" : [r],
                                     "prob"  : [len(filtered[filtered["correct"]])/float(len(filtered))]}))
    to_plot = to_plot.sort("ratio")
    plot_probs(to_plot["ratio"], to_plot["prob"], "", "Ratio of Correct Answer Before")


figure(figsize=(15,5))
plt.subplot(1, 2, 1)
plot_prob_correct_answer_by_column("question.number.all", "All Answers")
plot_prob_correct_answer_by_column("question.number.correct", "Correct Answers")
plot_prob_correct_answer_by_column("question.number.wrong", "Wrong Answers")
plt.subplot(1, 2, 2)
plot_prob_correct_answer_by_success()
plt.show()
```
