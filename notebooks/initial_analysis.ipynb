# Inital Analysis

```python
import sys
sys.path.append("../libs")

import geodata
import matplotlib.pyplot as plt

analysis = geodata.PlotAnalysis("../target/data/answers.csv")

plt.figure(figsize=(15, 10))

analysis.plot_pattern_times_density(
    pattern = [1,0],
    plot = [True, False],
    labels = ["Correct -> Wrong"],
    normalize = "user")

analysis.plot_pattern_times_density(
    pattern = [1,1],
    plot = [True, False],
    labels = ["Correct -> Correct"],
    normalize = "user")

plt.legend()
plt.show()
```

## Response Time

```python
plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
analysis.plot_time_hist(True);
plt.legend()
plt.subplot(1, 2, 2)
analysis.plot_time_hist(False);
plt.legend()
plt.show()
```

```python
analysis.plot_time_density(True)
analysis.plot_time_density(False)
plt.legend()
plt.show()
```

## Previous Correct Answer Response Time

```python
plt.figure(figsize=(15, 10))

analysis.plot_pattern_times_density(
    pattern = [1,0],
    plot = [True, False],
    labels = ["Correct -> Wrong"],
    normalize = "user")

analysis.plot_pattern_times_density(
    pattern = [1,1],
    plot = [True, False],
    labels = ["Correct -> Correct"],
    normalize = "user")

plt.legend()
plt.show()
```
